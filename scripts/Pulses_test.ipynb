```python
from CQEDModel import *
from Pulses import *
import qutip as qp
import time

dur=1.0 #Pulse duration
fm=100.0 #Frequency for deterministic noise form
na=0.05 #Amplitude of determinisitic noise (relative to pulse amplitude)

tlist=np.linspace(-3*dur,6*dur,400)
tlist0=np.linspace(-3*dur,6*dur,10000)

'''
Gaussian pulse profile with random noise introduce to envelope
'''

pulse1=Pulse(amplitude=1.0, frequency=0, phase=0, pulse_type="Gaussian", duration=dur, center=dur, noise=generate_pulse_noise\
             ,noiseamp=na,freqmult=fm,sq=False)
# Save timestamp
start = time.time()
pulse1.precomputing_noise(tlist0)
# Save timestamp
end = time.time()

print(end-start)
start = time.time()
pulse1.plot_pulse(tlist)
end = time.time()
print(end-start)
```
